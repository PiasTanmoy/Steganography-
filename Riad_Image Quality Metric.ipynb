# Peak Signal to Noise Ratio(PSNR):


```
def PSNR(img1, img2):
    #mse = numpy.mean( (img1 - img2) ** 2 )
    mse = sk_mse(img1, img2)
    if mse == 0:
      return 100
    M = 255.0
    return 10 * math.log10(M**2 / mse)
```

# Average Difference(AD)

```
def AD(A, B):
  return numpy.mean(A - B)
```

# Maximum Difference(MD):

```
def MD(A, B):
  return np.max(A-B)
```

# Mean Absolute Error(MAE): 

```
def MAE(A,B):
  sum = 0
  for i in range(M):
    for j in range(N):
      sum += np.abs(A[i][j] - B[i][j])

  MAE = (1/(M*N))*sum
  MAE = np.average(MAE)
  return MAE
```

# Peak Mean Square Error(PMSE)



```
def PMSE(A, B):
  mse = sk_mse(A, B)
  pmse = mse / np.max(A)**2
  return pmse
```

# Structural Similarity Index (SSIM)

```
def SSIM(A, B):
  return sk_ssim(A,B, multichannel=True)
```

# Structural Dissimilarity (DSSIM)

```
def DSSIM(A,B):
  ssim_xy = sk_ssim(A,B, multichannel=True)
  dssm = 1.0/ (1-ssim_xy)
  return dssm
```

# SNR

```
def SNR(img1, img2):
    mse = sk_mse(img1, img2)
    if mse == 0:
      return 100
    R2 = np.sum(img1**2)
    if R2 == 0.0:
      R2 = 0.00000000001
    #print(mse, R2, R2 / (mse*M*N) )
    return 10 * math.log10( R2 / (mse*M*N) )
```

# Structure Content

```
def SC(A,B):
  x = np.sum(A)**2
  y = np.sum(B)**2
  SC_val = x/y
  return SC_val
```

# N. Cross-Correlation

```
def NCC(A, B):
  return np.sum(A*B) / np.sum(A**2)
```

# Correlation Quality

```
def CQ(A, B):
  return np.sum(A*B) / np.sum(A)
```

# Image Fidelity

```
def IF(A,B):
  return (1 - ( np.sum((A-B)**2) / np.sum(A**2) ))
```
